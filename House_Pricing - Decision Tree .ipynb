```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv("house_pricing.csv")
df.head()
```

```python
df_ready = df.dropna(axis=1)
```

```python
df_ready.head()
```

```python
df.isnull().sum()
```

```python
X = df_ready.iloc[:, :-1]
y = df_ready.iloc[:, -1]
```

```python
X = pd.get_dummies(df_ready, drop_first = True) #retire la derniere colonne sinon colinearité
```

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred = pd.DataFrame(y_pred)
```

```python
y_pred.head()
```

```python
y_test = pd.DataFrame(y_test)
```

```python
#pour évaluer la qualité du model il faut voir si les valeurs de y_pred et y_test sont eloignées
```

```python
evaluation = np.abs(y_pred.values - y_test.values)  # en valeur absolue 
```

```python
pd.DataFrame(evaluation)  #ecart entre les deux valeurs
```

```python
#faire l'erreur moyenne au carré
```

```python
evaluation.mean()
```

```python
y.mean()
```

```python
evaluation.mean()/y.mean()*100
```

```python
from sklearn.ensemble import RandomForestRegressor   
regressor_rf = RandomForestRegressor(n_estimators = 20, random_state = 0)
regressor_rf.fit(X_train, y_train)
```

```python
y_pred_rf = regressor_rf.predict(X_test)
```

```python
pd.DataFrame(y_pred_rf).head()
```

```python
y_test.head()
```
